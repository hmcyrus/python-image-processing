## Task 1: Create a boxfilter with the size 5X5 and apply it on an image

Don't forget to normalize! use : filter_gaussian.png; 

**Solution starts here**

First we are reading and displaying the image

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

img = cv2.imread('filter_gaussian.png')

# showing image using opencv, this opens a new window
# cv2.imshow('GFG', img)
# cv2.waitKey(0)

# showing image using matplotlib, this prints the image in output
plt.imshow(img)
# print(img.shape)
img_height = img.shape[0]
img_width = img.shape[1]
```

Then we are applying a boxfilter with size 5x5 on the image we read above.

Reference: https://docs.opencv.org/4.x/d4/d13/tutorial_py_filtering.html

```python
kernel = np.ones((5,5),np.float32)/50  
# increasing the value of the denominator will make the image blurrier
# print(kernel)
img_after_applying_box_filter = cv2.filter2D(img,-1,kernel)

plt.imshow(img_after_applying_box_filter)

norm = np.zeros((img_width,img_height))
norm_image = cv2.normalize(img,norm,0,255,cv2.NORM_MINMAX)


plt.subplot(131)
plt.imshow(img)
plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(132)
plt.imshow(img_after_applying_box_filter)
plt.title('After Filtering')
plt.xticks([]), plt.yticks([])
plt.subplot(133),
plt.imshow(norm_image),
plt.title('Normalized')
plt.xticks([]), plt.yticks([])

plt.show()
```

# Task 2: Create a gaussianfilter with several sizes 3x3,7x7, 15x15

What happens to an image by using this several sizes.use : filter_gaussian.png; You can use : cv2.GaussianBlur(img,kernelsize,0)

Following is a demonstration of What happens after applying Gaussian filter of several sizes

## Task 3: Using a Median filter on a salt and pepper image

Use  cv2.medianBlur(img,kernelsize); Use these images: noisy_multiply; noisy_add;

```python
# saltandpepper_image = cv2.imread('saltandpepper_add.png')
saltandpepper_image = cv2.imread('saltandpepper_multiply.png')

img_width = saltandpepper_image.shape[0]
img_height = saltandpepper_image.shape[1]
median = cv2.medianBlur(saltandpepper_image,5)

norm = np.zeros((img_width,img_height))
norm_image = cv2.normalize(median,norm,0,255,cv2.NORM_MINMAX)


plt.subplot(131)
plt.imshow(saltandpepper_image)
plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(132)
plt.imshow(median)
plt.title('After Median Filtering')
plt.xticks([]), plt.yticks([])
plt.subplot(133),
plt.imshow(norm_image),
plt.title('Normalized')
plt.xticks([]), plt.yticks([])
plt.show()
```
