## Task 1: Create a boxfilter with the size 5X5 and apply it on an image

Don't forget to normalize! use : filter_gaussian.png; 

**Solution starts here**

First we are reading and displaying the image

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

img = cv2.imread('filter_gaussian.png')

# showing image using opencv, this opens a new window
# cv2.imshow('GFG', img)
# cv2.waitKey(0)

# showing image using matplotlib, this prints the image in output
plt.imshow(img)
```

Then we are applying a boxfilter with size 5x5 on the image we read above.

Reference: https://docs.opencv.org/4.x/d4/d13/tutorial_py_filtering.html

```python
kernel = np.ones((5,5),np.float32)/50  
# increasing the value of the denominator will make the image blurrier
# print(kernel)
img_after_applying_box_filter = cv2.filter2D(img,-1,kernel)

plt.imshow(img_after_applying_box_filter)
```

# Task 2: Create a gaussianfilter with several sizes 3x3,7x7, 15x15

What happens to an image by using this several sizes.use : filter_gaussian.png; You can use : cv2.GaussianBlur(img,kernelsize,0)

Following is a demonstration of What happens after applying Gaussian filter of several sizes 

```python
img_after_applying_gaussian_filter_3x3 = cv2.GaussianBlur(img,(3,3),0)
plt.imshow(img_after_applying_gaussian_filter_3x3)
```

```python
img_after_applying_gaussian_filter_7x7 = cv2.GaussianBlur(img,(7,7),0)
plt.imshow(img_after_applying_gaussian_filter_7x7)
```

```python
img_after_applying_gaussian_filter_15x15 = cv2.GaussianBlur(img,(15,15),0)
plt.imshow(img_after_applying_gaussian_filter_15x15)
```

**Finding** As the filter size in gaussian filter increases image gets blurrier
